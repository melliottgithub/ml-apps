## Text lowercase

```python
text = text.lower()
text
```

## Tokenization

```python
from nltk.tokenize import word_tokenize

words = word_tokenize(text)
words
```

## Initializing Stop words and stemmer

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

stop_words = set(stopwords.words('english'))
ps = PorterStemmer()
```

## Processing the words and removing stopwords

```python
tokens = [ps.stem(word) for word in words if word not in stop_words]
tokens[:5]
```

```python
processes_text = ' '.join(tokens)
processes_text
```

## word counter frequencies

```python
from collections import Counter

word_freq = Counter(processes_text.split())
word_freq
```

```python
terms_50_or_more = [term for term, freq in word_freq.items() if freq >= 50]
terms_10_or_more = [term for term, freq in word_freq.items() if freq >= 10]
print(terms_50_or_more)
print(terms_10_or_more)
```

## Full function -- removing punctucationas well

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

nltk.download('punkt')
nltk.download('stopwords')

def text_parse_nltk(text, lemmatize=False, keep_numbers=False):
    text = re.sub(r'[^a-zA-Z\s]' if not keep_numbers else r'[^a-zA-Z0-9\s]', '', text.lower())
    words = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    processor = WordNetLemmatizer() if lemmatize else PorterStemmer()
    result_tokens = [processor.lemmatize(word) if lemmatize else processor.stem(word) for word in words if word not in stop_words and word]
    preprocessed_text = ' '.join(result_tokens)
    return preprocessed_text

def analyze_text(text):

    preprocessed_text = text_parse_nltk(text, lemmatize=True, keep_numbers=False)
    
    book_words = word_tokenize(preprocessed_text)

    term_freq = Counter(book_words)

    frequent_terms = [term for term, freq in term_freq.items() if freq >= 50]
    common_terms = [term for term, freq in term_freq.items() if freq >= 10]
    
    wordcloud = WordCloud(width=800, height=400, background_color='white')
    wordcloud.generate(preprocessed_text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    
    return frequent_terms, common_terms

# Example usage:
book_text = chapters[1]
frequent_terms, common_terms = analyze_text(book_text)
print("Frequent Terms:", frequent_terms)
print("Common Terms:", common_terms)
```

## Summary

The code development begins with the creation of a text preprocessing function, text_parse_nltk from the previous week. 

This function handles essential tasks, including converting text to lowercase, eliminating non-alphabetical characters, word tokenization, stopword removal, and optional lemmatization or stemming. 

Following text preprocessing, the code proceeds to term frequency analysis through the analyze_text function. This function tokenizes the preprocessed text, removes punctuation, and employs the Counter class to compute term frequencies. It effectively identifies terms that occur at least 50 times and 10 times within the text, providing valuable insights into the content.

Finally, the code harnesses the WordCloud library to generate visually informative word cloud visualizations. These word clouds offer an intuitive and visually appealing representation of the most common terms within the text. 

Collectively, these steps culminate in the creation of a powerful text analysis tool with diverse applications, where understanding textual data is essential.
